```python
import pandas as pd

# separar los datos por tecnología
df_adsl = df_pivot['adsl']
df_fibra = df_pivot['fibra óptica']
df_cablemodem = df_pivot['cablemodem']
df_otros = df_pivot['otros']
df_wireless = df_pivot['wireless']

```

```python
df_adsl.head(2)
```

```python

# calcular los cambios porcentuales por cada columna
cambios = df_adsl.pct_change(axis=1)
```

```python
cambios.head(3)
```

# debo hacer la trasnpuesta de cada uno de los df y guardarlos de manera separada en diferentes csvs

```python
new_dfs = [df_adsl, df_fibra, df_cablemodem, df_otros, df_wireless]
cadena_string_df = ["df_adsl", "df_fibra", "df_cablemodem", "df_otros", "df_wireless"]

for i, df in enumerate(new_dfs):
    cambios_transposed = df.pct_change(axis=1).transpose()
    print(cambios_transposed.head(10))
    cambios_transposed.to_csv(f'../ETL/CSV Cleaned/{cadena_string_df[i]}.csv', mode='w', index=False, header=True)
```
