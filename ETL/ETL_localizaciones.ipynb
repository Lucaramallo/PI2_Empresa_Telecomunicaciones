```python
import pandas as pd

# Supongamos que su DataFrame se llama 'df'
df['Latitud'] = df['Latitud'].astype(float)
df['Longitud'] = df['Longitud'].astype(float)
```

```python
df.info()
```

```python
df['Latitud'] = df['Latitud'] / (10 ** 13)
df['Longitud'] = df['Longitud'] / (10 ** 13)
```

```python
df.head(50)
```

```python
df.loc[df['Latitud'] > -20.000, 'Latitud'] += df.loc[df['Latitud'] < -20.000, 'Latitud'].mean()
```

```python
df.loc[df['Longitud'] > -20.000, 'Longitud'] += df.loc[df['Longitud'] < -20.000, 'Longitud'].mean()
```

```python
import matplotlib.pyplot as plt

# Selecciona las columnas de latitud y longitud del DataFrame
latitud = df['Latitud']
longitud = df['Longitud']

# Crea una figura y un eje para el scatterplot
fig, ax = plt.subplots()

# Crea el scatterplot de las coordenadas
ax.scatter(longitud, latitud, alpha=0.5, s=5)

# Configura los títulos y etiquetas de los ejes
ax.set_title('Localizaciones')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

# Muestra el scatterplot
plt.show()
```
