Ac√° deber√≠a rellenar cada valor nulo con el mismo valor que tiene la fila. Y despu√©s multiplicarla por la. Matrix, norma. Revisada de arriba. Y as√≠ obtendr√≠a una muy pero muy, muy precisa, pero no exacta medida de todos los ingresos por provincia. En cuanto a. ¬øRealmente cu√°nto Representa de la porci√≥n de mercado de cada provincia al mercado argentino y cu√°nto genera cada una de ellas en termonos monetarios de INGRESOS?

```python
df_ingresos_por_periodo = df_ingresos_por_periodo.fillna(0)
```

```python
df_ingresos_por_periodo.info()
```

```python
df_ingresos_por_periodo = df_ingresos_por_periodo.drop([0, 1])
```

```python
df_ingresos_por_periodo = df_ingresos_por_periodo.fillna(0.0)
# definir una funci√≥n personalizada para encontrar el primer valor mayor a cero en una fila (excluyendo la primera columna)
def get_first_nonzero(row):
    return next((value for value in row.values[1:] if value > 0), 0)

# aplicar la funci√≥n personalizada a cada fila y agregar el resultado como una nueva columna
df_ingresos_por_periodo['total'] = df_ingresos_por_periodo.apply(get_first_nonzero, axis=1)

# imprimir el DataFrame resultante
df_ingresos_por_periodo.head(50)
```

```python
df_ingresos_por_periodo.head(50)
```

```python
cols = [c for c in df_ingresos_por_periodo.columns if c != 'total']

for i, row in df_ingresos_por_periodo.iterrows():
    for c in cols:
        if row[c] == 0:
            df_ingresos_por_periodo.loc[i, c] = row['total']

```

```python
df_ingresos_por_periodo.head(50)
```

```python
df_ingresos_por_periodo = df_ingresos_por_periodo.iloc[:, 1:-1]
df_ingresos_por_periodo = df_ingresos_por_periodo.reset_index(drop=True)
```

```python
df_ingresos_por_periodo.head(50)
```

```python
ingresos_x_provincia = df_acc_int_cada_100 * df_ingresos_por_periodo
ingresos_x_provincia = ingresos_x_provincia
```

```python
df_listado_provincias = pd.read_csv('./CSV Cleaned/listado_provincias.csv')
```

```python
df_listado_provincias
```

```python
provincias = df_listado_provincias.loc[:, "provincia"].iloc[1:].reset_index()
provincias = provincias.drop(columns='index')
```

```python
provincias.head(24)
```

```python
ingresos_x_provincia = pd.concat([provincias, ingresos_x_provincia], axis=1)
ingresos_x_provincia
```

```python
pivot_ingresos = pd.pivot_table(ingresos_x_provincia, index="provincia")
```

```python
pivot_ingresos
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='white')
fig, ax = plt.subplots(figsize=(20, 14))
sns.heatmap(pivot_ingresos.corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Crear un nuevo dataframe con los datos de los √∫ltimos tres meses y los pr√≥ximos tres meses para cada provincia
df_regresion = pivot_ingresos.iloc[:, -12:]

# Definir X e y
X = np.array(df_regresion.iloc[:, :-6])
y = np.array(df_regresion.iloc[:, -6:])

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

# Crear una instancia de LinearRegression y ajustar el modelo a los datos de entrenamiento
reg = LinearRegression().fit(X_train, y_train)

# Hacer predicciones para los pr√≥ximos tres meses utilizando los datos de prueba
predicciones = reg.predict(X_test)

# Imprimir las predicciones
print(predicciones)

```

```python
from sklearn.metrics import r2_score

# Calcular R2 en los datos de prueba
r2 = r2_score(y_test, predicciones)

# Imprimir el valor de R2
print("El valor de R2 es:", r2)
```

```python
# Hacer predicciones para los pr√≥ximos tres meses utilizando los datos de prueba
predicciones = reg.predict(X)

# Agregar las tres columnas de pron√≥stico al data frame pivot_ingresos
columnas_pronostico = ['2022.3', '2023', '2023.1','2023.2', '2023.3', '2024']
for i, col in enumerate(columnas_pronostico):
    pivot_ingresos[col] = predicciones[:, i]

# Imprimir el data frame actualizado
pivot_ingresos
```

```python
# Guardar el dataframe transpuesto en un archivo CSV
pivot_ingresos.to_csv('./CSV Cleaned/Ingresos_x_Provincia_en_miles_de_pesos.csv', mode='w', index=False, header=True)
```

```python
import matplotlib.pyplot as plt

# Obtener una lista de las provincias en el data frame
provincias = pivot_ingresos.index

# Configurar la figura
figsize = (10, 5)  # Tama√±o de la figura
title_fontsize = 20  # Tama√±o de la fuente del t√≠tulo
label_fontsize = 12  # Tama√±o de la fuente de las etiquetas

# Graficar los datos de ingresos reales y las predicciones para cada provincia
for prov in provincias:
    # Crear una nueva figura para cada provincia
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    fig.suptitle(f'Predicciones de ingresos para {prov}', fontsize=title_fontsize)

    # Graficar los datos de ingresos reales y las predicciones
    ax.plot(pivot_ingresos.columns, pivot_ingresos.loc[prov, :], label='Ingresos reales')
    ax.plot(columnas_pronostico, pivot_ingresos.loc[prov, columnas_pronostico], label='Ingresos pronosticados')
    ax.set_xlabel('Meses', fontsize=label_fontsize)
    ax.set_ylabel('Ingresos', fontsize=label_fontsize)
    ax.legend()
    plt.xticks(rotation=90)
    ax.grid(True)  # Activa la grilla en los ejes
    # Mostrar el gr√°fico para la provincia actual
    plt.show()


```
