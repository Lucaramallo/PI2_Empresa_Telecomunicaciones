```python
df.columns = df.columns.str.lower()
df = df.rename(columns={'año': 'anio'})
# Eliminar las comillas de la última columna y convertirla a float
df["accesos por cada 100 hogares"] = df["accesos por cada 100 hogares"].str.replace(",", ".").astype(float)

df['accesos por cada 100 hogares'] = pd.to_numeric(df['accesos por cada 100 hogares'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Seleccionar sólo las columnas numéricas para el boxplot
cols = [col for col in df.columns if col not in []]

# Crear un scatter plot separado para cada columna
for col in cols:
    fig, ax = plt.subplots(figsize=(6, 4))
    df.plot.scatter(x=col, y='accesos por cada 100 hogares', ax=ax,)
   
    ax.set_ylabel('accesos por cada 100 hogares')
    ax.set_xlabel(f'Scatter plot de {col}')
    ax.set_title(f'Scatter plot de {col}')
    if col == 'provincia':
        ax.set_xticklabels(df['provincia'], rotation=90)
    else:
        pass

    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar los datos por provincia
grupos = df.groupby('provincia')

# Iterar sobre cada grupo y hacer un gráfico de línea para cada columna
for nombre_provincia, grupo in grupos:
    # Obtener todas las columnas excepto 'provincia', 'anio' y 'trimestre'
    columnas = [col for col in grupo.columns if col not in ['provincia', 'anio', 'trimestre', 'variacion de los accesos por cada 100 hogares']]
    
    # Crear un gráfico de línea para cada columna
    for columna in columnas:
        # Crear un gráfico de línea suave para la columna
        x = grupo['anio'] + grupo['trimestre']/4
        y = grupo[columna].interpolate()
        plt.scatter(x, y, label=columna, linewidth=2, c='orange')
        
        # Calcular la tendencia usando una regresión lineal
        slope, intercept = np.polyfit(x, y, 1)
        trendline = intercept + slope*x
        
        # Crear un gráfico de línea punteada para la tendencia
        plt.plot(x, trendline, linestyle='--', color='blue', label='Tendencia')
        
        # Agregar la leyenda de la tendencia
        x_df = pd.DataFrame(x)
        trendline_df = pd.DataFrame(trendline)
        plt.text(x_df.iloc[-1]+0.5, trendline_df.iloc[-1], f'{columna} tendencia', color='blue')

        # Calcular la pendiente de la tendencia
        tendencia_coef = np.polyfit(x, y, 1)[0]
        # Agregar la leyenda de la tendencia con la pendiente
        plt.text(x.iloc[-1]+0.5, trendline.iloc[-1], f'{columna} tendencia (pendiente: {tendencia_coef:.2f})', color='blue')



    # Añadir etiquetas y leyendas
    plt.xlabel('Año')
    plt.ylabel('Valor')
    plt.title(f'Evolución de los accesos por cada 100 hogares en {nombre_provincia}')
    plt.legend()
    # Activar líneas de cuadrícula con estilo punteado de color gris claro
    plt.grid(linestyle='--', color='lightgrey')
    
    # Mostrar el gráfico
    plt.show()

```

```python
# ordenar el DataFrame por año y trimestre
df = df.sort_values(by=["anio", "trimestre"])

# calcular la variación porcentual entre los valores de accesos de un trimestre a otro
df["variacion de los accesos por cada 100 hogares"] = df.groupby("provincia")["accesos por cada 100 hogares"].pct_change()

# rellenar los valores nulos con ceros en la columna "variacion"
df["variacion de los accesos por cada 100 hogares"] = df["variacion de los accesos por cada 100 hogares"].fillna(0)

# mostrar el resultado
df.head(50)
```

```python
# Guardar DataFrame como archivo CSV
df.to_csv('./CSV Cleaned/Penetración de Internet fijo (accesos por cada 100 hogares) CLEANED.csv',mode='w', index=False)

```

# la transpuesta:

```python
# Obtener la transpuesta con las provincias en las filas y los valores Y agrupados en columnas de año y trimestre
df_transpuesta = df.pivot_table(index='provincia', columns=['anio', 'trimestre'], values='accesos por cada 100 hogares')

# Mostrar el dataframe transpuesto
df_transpuesta
```

```python

# Convertir el índice en una columna y crear un nuevo DataFrame
df_listado_provincias = df_transpuesta.reset_index()[['provincia']]

print(df_listado_provincias)

# Guardar el dataframe transpuesto en un archivo CSV
df_listado_provincias.to_csv('./CSV Cleaned/listado_provincias.csv', index=False, header=True, mode='w')
```

```python
# Guardar el dataframe transpuesto en un archivo CSV
df_transpuesta.to_csv('./CSV Cleaned/Penetración de Internet fijo (accesos por cada 100 hogares) _ Transpuesta.csv', index=False, header=True, mode='w')
```
