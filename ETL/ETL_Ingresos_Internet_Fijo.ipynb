```python
df.columns = df.columns.str.lower()
```

```python
df.rename(columns={'año': 'anio'}, inplace=True)
```

```python
df.head(5)
```

```python
df['periodo2'] = df.apply(lambda x: str(x['anio']) + str(x['trimestre']), axis=1)
df.drop(['periodo'], axis=1, inplace=True)
```

```python
df.rename(columns={'periodo2': 'periodo'}, inplace=True)
```

# transpuesta:

```python
# Hacer la transpuesta o pivot table
df_pivot = df.pivot(index='periodo', columns=['anio','trimestre'], values='ingresos (miles de pesos)')

df_pivot.head(25)
```

```python
df_pivot.fillna(0)
```

```python
# Guardar el dataframe transpuesto en un archivo CSV
df_pivot.to_csv('./CSV Cleaned/Ingresos_internet_fijo _ Transpuesta.csv', mode='w', index=False, header=True)
```
