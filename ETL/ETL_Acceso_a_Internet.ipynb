```python
import pandas as pd

df = pd.read_csv('../CSV data importada/CSV Accesos a internet/Acceso a Internet Fijo por rangos de velocidad de bajada y provincia.csv')
```

```python
df.isnull().sum()
```

```python
df['OTROS'].fillna(method='ffill', inplace=True)
df.drop('Total', axis=1, inplace=True)
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.iloc[:, [3, 4, 6, 7, 8, 10]] = df.iloc[:, [3, 4, 6, 7, 8, 10]].astype(float)
```

```python
# Calcular el promedio de los valores adyacentes para la columna "HASTA 512 kbps"
avg_values = df['HASTA 512 kbps'].rolling(window=3, min_periods=1).mean()

# Reemplazar los valores nulos con el promedio de los valores adyacentes
df['HASTA 512 kbps'] = df['HASTA 512 kbps'].fillna(avg_values)
df.iloc[:, [3, 4, 6, 7, 8, 10]] = df.iloc[:, [3, 4, 6, 7, 8, 10]].astype(float)
```

```python
df['OTROS'] = df['OTROS'].abs() #valores absolutos habia algunos negativos
```

```python
df.info()
```

```python

# Seleccionar sólo las columnas numéricas para convertir
cols = [col for col in df.columns if col != 'Provincia']

# Convertir los valores a float y reemplazar NaN con la media
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(df[col].mean(), inplace=True)

# Verificar los cambios
df.info()
```

```python
df.head(5)
```

```python
df = df.rename(columns={
    'Año': 'anio',
    'Provincia': 'provincia',
    'Trimestre': 'trimestre',
    'HASTA 512 kbps': 'hasta_512kbps',
    '+ 512 Kbps - 1 Mbps': '512kbps_1mbps',
    '+ 1 Mbps - 6 Mbps': '1mbps_6mbps',
    '+ 6 Mbps - 10 Mbps': '6mbps_10mbps',
    '+ 10 Mbps - 20 Mbps': '10mbps_20mbps',
    '+ 20 Mbps - 30 Mbps': '20mbps_30mbps',
    '+ 30 Mbps': 'mas_30mbps',
    'OTROS': 'otros'
})
```

```python
# Crear una nueva columna con la suma de todas las columnas
df['total'] = df[['hasta_512kbps', '512kbps_1mbps', '1mbps_6mbps', '6mbps_10mbps', '10mbps_20mbps', '20mbps_30mbps', 'mas_30mbps', 'otros']].sum(axis=1)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

numeric_cols = ['anio','hasta_512kbps','512kbps_1mbps','6mbps_10mbps','10mbps_20mbps','20mbps_30mbps','otros', 'total']

data_numeric = df[numeric_cols]

# Convert non-numeric data to NaN values
data_numeric = data_numeric.apply(pd.to_numeric)

corr = data_numeric.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Seleccionar sólo las columnas numéricas para el boxplot
cols = [col for col in df.columns if col not in ['provincia', 'anio', 'trimestre']]

# Crear un scatter plot separado para cada columna
for col in cols:
    fig, ax = plt.subplots(figsize=(6, 4))
    df.plot.scatter(x='provincia', y=col, ax=ax)
    ax.set_ylabel('Velocidad de Internet (Mbps)')
    ax.set_xlabel('Provincia')
    ax.set_title(f'Scatter plot de {col}')
    
    # Rotar las etiquetas del eje x
    ax.set_xticklabels(df['provincia'], rotation=90)

    plt.show()
```

```python
# Seleccionar sólo las columnas numéricas
cols = [col for col in df.columns if col != 'provincia']

# Crear un histograma para cada columna
for col in cols:
    fig, ax = plt.subplots(figsize=(6, 4))
    df.hist(column=col, ax=ax, bins=10)
    ax.set_xlabel(col)
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'Histograma de {col}')
    plt.show()
```
